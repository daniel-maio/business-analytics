```python
import pandas as pd
import numpy as np
import ast
```

# Load Data

```python
with open(file='./data/frases.txt', mode='r', encoding='utf-8') as file:
    content = file.read()
```

- ast.parse(txt) analisa estrutura sintática.
- se o txt está em uma linguagem que o Python entende, ele lê como se fosse linguagem Python

```python
tree = ast.parse(content)
```

* Objeto ast.Assign é um node
* Ele gerou um objeto Node para cada atribuição que encontrou no txt.
* São 3 atribuições, então são 3 objetos Assign.
* assignemnt → variável = alguma coisa

```python
tree.body
```

```python
tree.body[0]
```

- Cada nó tem 2 atributos (targets e value)
- Target: nome da variável
- Value: lista

```python
target_node_0 = tree.body[0].targets[0].id
target_node_0
```

```python
value_node_0 = ast.literal_eval(tree.body[0].value)
value_node_0
```

- O atributo value possui o atributo ELTS que é uma lista de elementos (constantes).
- Cada constante é uma linha de uma Lista

```python
tree.body[0].value.elts
```

```python
def gera_dados(arvore):
    '''Extrai nome da lista e valores da lista
    de um arquivo txt'''
    dados = []

    for node in arvore.body:
        nome = node.targets[0].id
        
        for row in node.value.elts:
            dados.append({'label': nome,
                          'value': row.value})
    return dados
```

```python
data = gera_dados(tree)
```

```python
type(data)
```

```python
data
```

```python
n = len(data)
n
```

```python
# Cria uma coluna com datas aleatórias
dates = pd.date_range(start='2020-01-01', periods = n)
```

# Cria o Dataframe

```python
df = pd.DataFrame(data = {'id': range(n),
                          'data': dates,
                          'texto': [d['value'] for d in data],
                          'sentimento': [d['label'] for d in data]}
)
```

```python
df
```

```python
df['sentimento'] = df['sentimento'].map({'textos_positivos': 'positivo',
                                        'textos_negativos': 'negativo',
                                        'textos_neutros': 'neutro'}
)
```
