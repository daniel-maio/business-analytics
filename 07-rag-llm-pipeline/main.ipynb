# Imports and Settings

```python
from llama_index.llms.groq import Groq
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core import Settings

from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Document
from llama_index.core.text_splitter import TokenTextSplitter
```

```python
import logging
logging.basicConfig(level=logging.WARNING)
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
GROQ_API_KEY = os.getenv("GROQ_API_KEY")
```

```python
Settings.llm = Groq(model="openai/gpt-oss-120b", api_key=GROQ_API_KEY)
```

```python
Settings.embed_model = HuggingFaceEmbedding(
    model_name="sentence-transformers/all-MiniLM-L6-v2")
```

# Pipeline

```python
reader = SimpleDirectoryReader(input_dir="./data", recursive=True) #FAQ1.docx e FAQ2.docx
```

```python
docs = reader.load_data()  # lista de objetos Document
len(docs)
```

```python
# 2. Extrair o texto de cada Document
texts = [doc.get_content() for doc in docs]
len(texts)
```

```python
# 3. Split
splitter = TokenTextSplitter(chunk_size=800, chunk_overlap=400)
```

```python
split_texts = splitter.split_texts(texts)
len(split_texts)
```

```python
# 4. Document objects
split_docs = [Document(text=t) for t in split_texts]
```

```python
# 5. Criar índice vetorial
index = VectorStoreIndex.from_documents(
    documents=split_docs,
    show_progress=True)
```

```python
# Cria um query engine a partir do índice
query_engine = index.as_query_engine(
    similarity_top_k=3)
```

# Query 1

```python
query_1 = "What are the payment methods"
```

```python
response_1 = query_engine.query(query_1)
print(response_1.response)
```

# Query 2


```python
query_2 = "Do the products come with a warranty"
```

```python
response_2 = query_engine.query(query_2)
print(response_2.response)
```
